# **Cleaning**

```python
houses.duplicated().sum()
```

```python
houses = houses.drop_duplicates()
houses.shape
```

```python
houses.drop(columns=['link','property_id'], inplace=True)
```

```python
houses.isnull().sum()
```

```python
houses['society'].value_counts()
```

```python
houses['society'] = houses['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
houses['society'] = houses['society'].str.replace('nan','independent')
```

```python
houses['society'].value_counts().shape
```

```python
houses['price'].value_counts()
```

```python
# Removing rows where price is state as 'Price on Request'
price_outlier = np.where(houses['price'] == 'Price on Request')
houses.drop(index=price_outlier[0], inplace=True)
```

```python
# Converting price into decimal
def price(x):
  if type(x) == float:
    return x
  else:
    if x[1] == 'Lac':
      return round(float(x[0])/100,2)
    else:
      return round(float(x[0]),2)

houses['price'] = houses['price'].str.split(' ').apply(price)
```

```python
houses.head(2)
```

```python
# Cleaning area column
houses.rename(columns={'rate' : 'price_per_sqft'}, inplace=True)
```

```python
houses['price_per_sqft'] = houses['price_per_sqft'].str.split('/').str.get(0).str.replace('₹ ','').str.replace(',','').astype(float)
```

```python
houses.head(3)
```

```python
# Cleaning Bedrooms column
houses['bedRoom'].value_counts()
```

```python
houses['bedRoom'].isnull().sum()
```

```python
# Removing rows where bedrooms is null
houses = houses[~houses['bedRoom'].isnull()]
houses.shape
```

```python
houses['bedRoom'] = houses['bedRoom'].str.split(' ').str.get(0).astype(int)
houses.head(3)
```

```python
# Doing same cleaning process for bathroom and balcony
houses['bathroom'].value_counts()
```

```python
houses['bathroom'].isnull().sum()
```

```python
houses['bathroom'] = houses['bathroom'].str.split(' ').str.get(0).astype(int)
houses.head(2)
```

```python
houses['balcony'].value_counts()
```

```python
houses['balcony'].isnull().sum()
```

```python
houses['balcony'] = houses['balcony'].str.split(' ').str.get(0).str.replace('No Balcony','0').str.replace('+','')
houses.head(3)
```

```python
# Cleaning Additional room columns
houses['additionalRoom'].value_counts()
```

```python
houses['additionalRoom'].isnull().sum()
```

```python
houses['additionalRoom'].fillna("not available",inplace=True)
houses['additionalRoom'] = houses['additionalRoom'].str.lower()
houses.head(3)
```

```python
# Cleaning floor number column
houses['noOfFloor'].value_counts()
```

```python
houses['noOfFloor'] = houses['noOfFloor'].str.split(' ').str.get(0)
```

```python
houses.rename(columns={'noOfFloor':'floorNum'},inplace=True)
houses.head()
```

```python
houses['facing'].fillna('NA',inplace=True)
```

```python
houses['area'] = round((houses['price']*10000000)/houses['price_per_sqft'])
```

```python
houses.insert(loc=1,column='property_type',value='house')
```
