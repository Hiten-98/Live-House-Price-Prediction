```python
flats = pd.read_csv('/content/drive/MyDrive/MTech/Project/flats.csv')
flats.head()
```

# **Cleaning**

```python
flats.drop(columns=['link','property_id'], inplace=True)
```

```python
flats.info()
```

```python
flats.isnull().sum()
```

```python
flats['society'].value_counts()
```

```python
flats['society'] = flats['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
flats['society'].value_counts()
```

```python
flats['price'].value_counts()
```

```python
# Removing rows where price is state as 'Price on Request'
price_outlier = np.where(flats['price'] == 'Price on Request')
flats.drop(index=price_outlier[0], inplace=True)
```

```python
# Converting price into decimal
def price(x):
  if type(x) == float:
    return x
  else:
    if x[1] == 'Lac':
      return round(float(x[0])/100,2)
    else:
      return round(float(x[0]),2)

flats['price'] = flats['price'].str.split(' ').apply(price)
```

```python
flats.head(2)
```

```python
# Cleaning area column
flats.rename(columns={'area' : 'price_per_sqft'}, inplace=True)
```

```python
flats['price_per_sqft'] = flats['price_per_sqft'].str.split('/').str.get(0).str.replace('₹ ','').str.replace(',','').astype(float)

# x = '₹ 6,224/sq.ft.'
# print(float(x.split('/')[0].replace('₹ ','').replace(',','')))
```

```python
flats.head(2)
```

```python
# Cleaning Bedrooms column
flats['bedRoom'].value_counts()
```

```python
flats['bedRoom'].isnull().sum()
```

```python
# Removing rows where bedrooms is null
flats = flats[~flats['bedRoom'].isnull()]
```

```python
flats['bedRoom'] = flats['bedRoom'].str.split(' ').str.get(0).astype(int)
flats.head(2)
```

```python
# Doing same cleaning process for bathroom and balcony
flats['bathroom'].value_counts()
```

```python
flats['bathroom'].isnull().sum()
```

```python
flats['bathroom'] = flats['bathroom'].str.split(' ').str.get(0).astype(int)
flats.head(2)
```

```python
flats['balcony'].value_counts()
```

```python
flats['balcony'].isnull().sum()
```

```python
flats['balcony'] = flats['balcony'].str.split(' ').str.get(0).str.replace('No Balcony','0').str.replace('+','')
flats.head(2)
```

```python
# Cleaning Additional room columns
flats['additionalRoom'].value_counts()
```

```python
flats['additionalRoom'].isnull().sum()
```

```python
flats['additionalRoom'].fillna("not available",inplace=True)
flats['additionalRoom'] = flats['additionalRoom'].str.lower()
flats.head(3)
```

```python
# Cleaning floor number column
flats['floorNum'].value_counts()
```

```python
flats['floorNum'] = flats['floorNum'].str.split(' ').str.get(0).str.replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.replace('st','').str.replace('nd','').str.replace('rd','').str.replace('th','')
```

```python
# Cleaning facing columns
flats['facing'].value_counts()
```

```python
flats['facing'].isnull().sum()
```

```python
flats['facing'].fillna('NA',inplace=True)
```

```python
flats.insert(loc=4,column='area',value=round((flats['price']*10000000)/flats['price_per_sqft']))
flats.insert(loc=1,column='property_type',value='flat')
```
